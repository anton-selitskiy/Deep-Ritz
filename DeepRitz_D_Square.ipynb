```python
import torch
import torch.nn as nn
#import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
b=0.5
```

```python
def exact_solution(x):
  return ((1/np.pi**2 + (b-1)/np.pi**2*(torch.sinh(2*np.pi*(1-x[:,0]/2)) + torch.sinh(np.pi*x[:,0]))/(np.sinh(2*np.pi)-2*b*np.sinh(np.pi)))*torch.sin(np.pi*x[:,1])).unsqueeze(-1)
```

```python
def reconstr(f, x):
  g = f(x)
  x1 = torch.zeros_like(x, device=x.device)
  x1[:, 0] = 1.0
  g[x[:,0]<=1] -= b*f(x[x[:,0]<=1]+x1[x[:,0]<=1])
  g[x[:,0]>1] -= b*f(x[x[:,0]>1]-x1[x[:,0]>1])
  return g/(1-b**2)
def reconstr_2D(x):
  return reconstr(exact_solution, x)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Parameters
#b = 1.0  # Example value for b
grid_size = 1000  # Number of points in each dimension

# Generate a grid of points
x0 = torch.linspace(0, 2, grid_size)
x1 = torch.linspace(0, 1, grid_size)
x0, x1 = torch.meshgrid(x0, x1)
x0 = x0.flatten()
x1 = x1.flatten()
points = torch.stack([x0, x1], dim=1)

# Evaluate the function at the grid points
z = reconstr_2D(points) # exact_solution(points)

# Reshape the results to match the grid
z = z.view(grid_size, grid_size).detach().numpy()
x0 = x0.view(grid_size, grid_size).numpy()
x1 = x1.view(grid_size, grid_size).numpy()

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x0, x1, z, cmap='viridis')

# Labels and title
ax.set_xlabel('x0')
ax.set_ylabel('x1')
ax.set_zlabel('exact_solution')
ax.set_title('Surface Plot of exact_solution')

plt.show()
```

```python
def right_hand_side(x):
  return torch.sin(np.pi*x[:,1]).unsqueeze(-1)
```

```python
class ResidualBlock(nn.Module):
    def __init__(self, hidden_size):
        super(ResidualBlock, self).__init__()
        self.fc = nn.Linear(hidden_size, hidden_size)
        self.relu = nn.ReLU()

    def forward(self, x):
        residual = x
        out = self.fc(x)
        out = self.relu(out)
        out = out + residual
        return out
```

```python
class ResidualModel(nn.Module):
    def __init__(self, input_size=2, hidden_size=20, output_size=1, num_layers=4):
        super(ResidualModel, self).__init__()
        self.input_layer = nn.Linear(input_size, hidden_size)
        self.hidden_layers = nn.ModuleList([ResidualBlock(hidden_size) for _ in range(num_layers)])
        self.output_layer = nn.Linear(hidden_size, output_size)
        self.relu = nn.ReLU()

    def forward(self, x):
        out = self.input_layer(x)
        out = self.relu(out)
        for layer in self.hidden_layers:
            out = layer(out)
        out = self.output_layer(out)
        return out
```

```python
input_size = 2
hidden_size = 20
output_size = 1
num_layers = 50

model = ResidualModel(input_size, hidden_size, output_size, num_layers)
```

```python
# Define the function FQ
def FQ(f, x):
    mask = (x[:,0] >=0) & (x[:,0]<=2) & (x[:,1]>= 0) & (x[:,1]<=1)
    result = torch.zeros((x.shape[0],1), device=x.device)
    result[mask] = (x[mask,0]*(x[mask,0]-2)*x[mask,1]*(x[mask,1]-1)).unsqueeze(-1)*f(x[mask])
    return result

# Define the function RQ
def RQ(fq, model1, x, b=0.5):
    x1 = torch.zeros_like(x, device=x.device)
    x1[:, 0] = 1.0  # Setting x1[:,0] to 1, for the shifting operations
    return fq(model1, x) + b * fq(model1, x + x1) + b * fq(model1, x - x1)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Parameters
#b = 1.0  # Example value for b
grid_size = 100  # Number of points in each dimension

# Generate a grid of points
x0 = torch.linspace(0, 2, grid_size)
x1 = torch.linspace(0, 1, grid_size)
x0, x1 = torch.meshgrid(x0, x1)
x0 = x0.flatten()
x1 = x1.flatten()
points = torch.stack([x0, x1], dim=1)

# Evaluate the function at the grid points
z = RQ(FQ, reconstr_2D, points) # exact_solution(points)
f = torch.zeros_like(z)
#print(z.shape)
mask = points[:,0] >= 1
sh = torch.zeros_like(points[mask])
sh[:,0] = 1
sh1 = torch.zeros_like(points[~mask])
sh1[:,0] = 1
f[mask] = (exact_solution(points[mask]) -0.5*exact_solution(points[mask]-sh))/(1-0.5**2)
f[~mask] = (exact_solution(points[~mask]) -0.5*exact_solution(points[~mask]+sh1))/(1-0.5**2)
# Reshape the results to match the grid
z = z.view(grid_size, grid_size).detach().numpy()
f = f.view(grid_size, grid_size).detach().numpy()
x0 = x0.view(grid_size, grid_size).numpy()
x1 = x1.view(grid_size, grid_size).numpy()

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x0, x1, z, cmap='viridis')

# Labels and title
ax.set_xlabel('x0')
ax.set_ylabel('x1')
ax.set_zlabel('exact_solution')
ax.set_title('Surface Plot of exact_solution')

plt.show()
```

```python
optimizer = optim.Adam(model.parameters(), lr=0.0001)
#optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

# Set up the learning rate scheduler to decrease the learning rate by a factor of 0.1 every 100 epochs
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=100, gamma=0.1)
```

```python
batch_size = 8*1024
for epoch in range(10000):
    model.train()
    scale = torch.ones((batch_size,2))
    scale[:,0] = 2*scale[:,0]
    x0 = torch.rand(batch_size, input_size, dtype=torch.float32)*scale
    x0.requires_grad = True
    gradients_1 = torch.autograd.grad(outputs=RQ(FQ, model, x0).sum(), inputs=x0, create_graph=True)[0]
    gradients_2 = torch.autograd.grad(outputs=((x0[:,0]*(x0[:,0]-2)*x0[:,1]*(x0[:,1]-1)).unsqueeze(-1)*model(x0)).sum(), inputs=x0, create_graph=True)[0]
    #loss = torch.mean((gradients - 2*right_hand_side(x0))*(x0[:,0]*(x0[:,0]-2)*x0[:,1]*(x0[:,1]-1)).unsqueeze(-1)*model(x0))
    slopes = torch.sum(gradients_1*gradients_2, dim=-1).unsqueeze(-1)
    #slopes = torch.square(gradients[:,0]) + torch.square(gradients[:,1])
    loss = torch.mean(slopes - 2*(x0[:,0]*(x0[:,0]-2)*x0[:,1]*(x0[:,1]-1)).unsqueeze(-1)*model(x0)*right_hand_side(x0))
    #loss = torch.mean(slopes - 2*RQ(FQ, model, x0)*right_hand_side(x0))
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    #scheduler.step()
    if epoch%100==0:
        model.eval()
        with torch.no_grad():
            x1 = torch.linspace(0, 2, steps=200)
            x2 = torch.linspace(0, 1, steps=100)
            x0 = torch.cartesian_prod(x1, x2)
            x0 = torch.rand(batch_size, input_size, dtype=torch.float32)
            u = (x0[:,0]*(x0[:,0]-2)*x0[:,1]*(x0[:,1]-1)).unsqueeze(-1)*model(x0)
            u_predict = reconstr_2D(x0)
            print(torch.mean(torch.abs(u - u_predict)))

```
